## Load some data

```python
valid_start = datetime(2017, 8, 21, 10)
valid_end   = datetime(2017, 8, 21, 11)

pc_data = ed.load_sp_eng_data('PC01A', valid_start, valid_end)
sc_data = ed.load_sp_eng_data('SC01A', valid_start, valid_end)

pc_data = pc_data.set_index('timestamp')
sc_data = sc_data.set_index('timestamp')
```

```python
print(sc_data.columns)
print(sc_data)
```

```python
f, ax = plt.subplots(2, sharex=True, figsize=(15,15))

ax[0].plot(pc_data.index, pc_data['roll'], '.',
              pc_data.index, pc_data['pitch'], '.',
                pc_data.index, pc_data['yaw'], '.')

ax[1].plot(sc_data.index, sc_data['roll'], '.',
          sc_data.index, sc_data['pitch'], '.',
          sc_data.index, sc_data['yaw'], '.')

```
